```
df = pd.read_csv('new_yelp.csv')
```

```
label_numeric = []
for i in range(df.shape[0]):
  if df['label'][i] == 'positive':
    label_numeric.append(1)
  else:
    label_numeric.append(0)
df['label_numeric'] = label_numeric
```
